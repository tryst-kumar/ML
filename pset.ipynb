```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Book1.csv')
```

```python
df.head()
```

```python
d = {'yes': 1, 'no': 0}
df['Give Birth'] = df['Give Birth'].map(d)
```

```python
d = {'no': 0, 'yes': 1}
df['Can Fly'] = df['Can Fly'].map(d)
```

```python
d = {'no': 0, 'yes': 1, 'sometimes': 2}
df['Live in Water'] = df['Live in Water'].map(d)
```

```python
d = {'no': 0, 'yes': 1}
df['Have Legs'] = df['Have Legs'].map(d)
```

```python
d = {'mammals': 1, 'non-mammals': 0}
df['Class'] = df['Class'].map(d)
```

```python
df
```

```python
features = ['Give Birth', 'Can Fly', 'Live in Water', 'Have Legs']
X = df[features]
y = df['Class']
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
```

```python
from sklearn import tree
tree.plot_tree(dtree, feature_names = features)
```

```python
def predicta():
    x = 1 if input('Enter if it can give birth (yes/no): ') == 'yes' else 0
    y = 1 if input('Enter if it can fly (yes/no): ') == 'yes' else 0
    z = {'yes': 1, 'no': 0}.get(input('Enter if it lives in water (yes/no/sometimes): '), 2)
    w = 1 if input('Enter if it has legs (yes/no): ') == 'yes' else 0

    t = dtree.predict([[x, y, z, w]])
    print('mammals') if t == 1 else print('non-mamamls')

```
